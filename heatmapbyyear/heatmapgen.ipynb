```python
import folium
import pandas as pd
from folium.plugins import HeatMap
```

```python
data = pd.read_csv('firedataseasons.csv')
```

```python
grouped_data = data.groupby('year')
```

```python
for year, group in grouped_data:
    # Create a new map for each year
    m = folium.Map(location=[28.3949, 84.1240], zoom_start=7)
    
    # Add heat layer to the map
    heat_data = group[['latitude', 'longitude']].values
    HeatMap(heat_data, radius=10).add_to(m)
    
    # Display the map
    print(year)
    display(m)
```
